```python
%run -n main.py
for item in items:
    if item.get('answer'):
        continue

    print(item)
    prompt = item_prompt(item)
    answers = []
    for _ in range(3):
        answer = ''
        async for chunk in yagpt_call(prompt):
            print(chunk, end='')
            answer += chunk
        answers.append(answer)
        print('\n---')
    item['answer'] = max(answers, key=len)
    print('\n\n')
```

# "такие темы я не обсуждаю"

```python
path = f'data/vicuna_yagpt_answers.jsonl'
lines = read_lines(path)
items = parse_jsonl(lines)

total, count = 0, 0
for item in items:
    text = item['answer']
    total += 1
    if (
        'такие темы я не обсуждаю' in text
        or 'говорить на эту тему я совсем не готова' in text
        or 'не могу ответить на этот вопрос' in text
        or 'Давайте поговорим о чем-то другом' in text
    ):
        print(item)
        count += 1

count, total, count / total
```
