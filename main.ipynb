```python
%run -n main.py
for item in items:
    if item.get('answer'):
        continue

    print(item)
    prompt = item_prompt(item)
    answers = []
    for _ in range(3):
        answer = ''
        async for chunk in yagpt_call(prompt):
            print(chunk, end='')
            answer += chunk
        answers.append(answer)
        print('\n---')
    item['answer'] = max(answers, key=len)
    print('\n\n')
```

```python
path = f'data/vicuna_yagpt_answers.jsonl'
lines = read_lines(path)
items = parse_jsonl(lines)

total, count = 0, 0
for item in items:
    text = item['answer']
    total += 1
    if (
        '—Ç–∞–∫–∏–µ —Ç–µ–º—ã —è –Ω–µ –æ–±—Å—É–∂–¥–∞—é' in text
        or '–≥–æ–≤–æ—Ä–∏—Ç—å –Ω–∞ —ç—Ç—É —Ç–µ–º—É —è —Å–æ–≤—Å–µ–º –Ω–µ –≥–æ—Ç–æ–≤–∞' in text
        or '–Ω–µ –º–æ–≥—É –æ—Ç–≤–µ—Ç–∏—Ç—å –Ω–∞ —ç—Ç–æ—Ç –≤–æ–ø—Ä–æ—Å' in text
        or '–î–∞–≤–∞–π—Ç–µ –ø–æ–≥–æ–≤–æ—Ä–∏–º –æ —á–µ–º-—Ç–æ –¥—Ä—É–≥–æ–º' in text
    ):
        print(item)
        count += 1

count, total, count / total
```
